```python
!pip -q install statsmodels scikit-learn tqdm

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
from pathlib import Path
from tqdm.auto import tqdm

from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("ARIMA setup completed.")
```

```python
ASSETS = {
    "BTC": "BTC_close",
    "ETH": "ETH_close",
    "BNB": "BNB_close"
}

HORIZONS = [1, 7, 30]

def compute_forecast_metrics(y_true, y_pred, origin_price):
    """
    Metrics:
    - MAE, RMSE, MAPE, R2
    - Directional Accuracy matet kundrejt origin_price:
      nëse modeli parashikon drejtimin e saktë nga origin deri te target.
    """
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    origin_price = np.asarray(origin_price, dtype=float)

    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / np.clip(np.abs(y_true), 1e-8, None))) * 100
    r2 = r2_score(y_true, y_pred)

    true_direction = np.sign(y_true - origin_price)
    pred_direction = np.sign(y_pred - origin_price)
    directional_accuracy = np.mean(true_direction == pred_direction) * 100

    return {
        "MAE": round(float(mae), 2),
        "RMSE": round(float(rmse), 2),
        "MAPE": round(float(mape), 2),
        "R2": round(float(r2), 4),
        "Directional_Accuracy": round(float(directional_accuracy), 2)
    }

print("Metrics function ready.")
```

```python
def select_arima_order_by_aic(train_series, max_p=5, max_q=3, d_values=[1]):
    """
    Zgjedh ARIMA(p,d,q) sipas AIC vetëm mbi Train.
    """
    best_order = None
    best_aic = np.inf
    results = []

    for p in range(max_p + 1):
        for d in d_values:
            for q in range(max_q + 1):
                order = (p, d, q)
                try:
                    model = ARIMA(
                        train_series,
                        order=order,
                        enforce_stationarity=False,
                        enforce_invertibility=False
                    )
                    fitted = model.fit()
                    aic = fitted.aic

                    results.append({
                        "order": order,
                        "aic": aic
                    })

                    if aic < best_aic:
                        best_aic = aic
                        best_order = order

                except Exception:
                    continue

    if best_order is None:
        raise ValueError("No valid ARIMA order found.")

    return best_order, best_aic, pd.DataFrame(results)


arima_orders = []
arima_order_details = []

for asset, target_col in ASSETS.items():
    print(f"\nSelecting ARIMA order for {asset}...")

    train_series = train_df.set_index("date")[target_col].astype(float)

    best_order, best_aic, details = select_arima_order_by_aic(
        train_series=train_series,
        max_p=5,
        max_q=3,
        d_values=[1]
    )

    print(f"{asset} best order: {best_order}, AIC: {best_aic:.2f}")

    arima_orders.append({
        "asset": asset,
        "target_col": target_col,
        "best_order": str(best_order),
        "best_aic": round(float(best_aic), 2)
    })

    details["asset"] = asset
    arima_order_details.append(details)

arima_orders_df = pd.DataFrame(arima_orders)
arima_order_details_df = pd.concat(arima_order_details, ignore_index=True)

display(arima_orders_df)

arima_orders_df.to_csv(RESULTS_DIR / "arima_orders.csv", index=False)
arima_order_details_df.to_csv(RESULTS_DIR / "arima_order_search_details.csv", index=False)

print("ARIMA order files saved.")
```

```python
def arima_walk_forward_forecast(train_val_series, test_series, order, horizon):
    """
    Walk-forward forecasting:
    - Modeli trajnohet fillimisht mbi Train + Validation.
    - Për çdo hap në Test, bëhet forecast për horizon ditë përpara.
    - Pas çdo hapi, modeli përditësohet me vlerën reale të re të Test,
      por pa përdorur target-in e ardhshëm.
    """
    train_val_series = train_val_series.astype(float)
    test_series = test_series.astype(float)

    fitted = ARIMA(
        train_val_series,
        order=order,
        enforce_stationarity=False,
        enforce_invertibility=False
    ).fit()

    records = []

    for i in range(0, len(test_series) - horizon + 1):
        forecast = fitted.forecast(steps=horizon)
        predicted_value = float(forecast.iloc[-1])

        if i == 0:
            origin_date = train_val_series.index[-1]
            origin_price = float(train_val_series.iloc[-1])
        else:
            origin_date = test_series.index[i - 1]
            origin_price = float(test_series.iloc[i - 1])

        target_date = test_series.index[i + horizon - 1]
        actual_value = float(test_series.iloc[i + horizon - 1])

        records.append({
            "date": target_date,
            "origin_date": origin_date,
            "asset": None,
            "horizon": horizon,
            "actual": actual_value,
            "origin_price": origin_price,
            "arima_pred": predicted_value
        })

        # Përditëso modelin me observimin real të sapo kaluar
        if i < len(test_series) - horizon:
            new_observation = test_series.iloc[[i]]
            try:
                fitted = fitted.append(new_observation, refit=False)
            except Exception:
                updated_history = pd.concat([train_val_series, test_series.iloc[:i+1]])
                fitted = ARIMA(
                    updated_history,
                    order=order,
                    enforce_stationarity=False,
                    enforce_invertibility=False
                ).fit()

    return pd.DataFrame(records)


train_indexed = train_df.set_index("date")
val_indexed = val_df.set_index("date")
test_indexed = test_df.set_index("date")

arima_predictions_all = []
arima_metrics_all = []

for _, row in arima_orders_df.iterrows():
    asset = row["asset"]
    target_col = row["target_col"]
    order = eval(row["best_order"])

    print(f"\nRunning ARIMA forecasts for {asset}, order={order}")

    train_val_series = pd.concat([
        train_indexed[target_col],
        val_indexed[target_col]
    ])

    test_series = test_indexed[target_col]

    for horizon in HORIZONS:
        print(f"  Horizon H={horizon}")

        pred_df = arima_walk_forward_forecast(
            train_val_series=train_val_series,
            test_series=test_series,
            order=order,
            horizon=horizon
        )

        pred_df["asset"] = asset
        pred_df["model"] = "ARIMA"

        metrics = compute_forecast_metrics(
            y_true=pred_df["actual"],
            y_pred=pred_df["arima_pred"],
            origin_price=pred_df["origin_price"]
        )

        metrics_row = {
            "model": "ARIMA",
            "asset": asset,
            "horizon": horizon,
            "n_predictions": len(pred_df),
            **metrics
        }

        arima_predictions_all.append(pred_df)
        arima_metrics_all.append(metrics_row)

arima_predictions_df = pd.concat(arima_predictions_all, ignore_index=True)
arima_metrics_df = pd.DataFrame(arima_metrics_all)

display(arima_metrics_df)

arima_predictions_df.to_csv(RESULTS_DIR / "arima_predictions.csv", index=False)
arima_metrics_df.to_csv(RESULTS_DIR / "arima_metrics.csv", index=False)

print("ARIMA predictions and metrics saved.")
print("Predictions:", RESULTS_DIR / "arima_predictions.csv")
print("Metrics:", RESULTS_DIR / "arima_metrics.csv")
```

```python
import random
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

SEED = 42

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

LOOKBACK = 60
BATCH_SIZE = 64
EPOCHS = 80
PATIENCE = 12
LR = 1e-3
HIDDEN_SIZE = 128
NUM_LAYERS = 2
DROPOUT = 0.2

FEATURE_COLS = [c for c in df.columns if c != "date"]
TARGET_COLS = ["BTC_close", "ETH_close", "BNB_close"]
TARGET_IDXS = [FEATURE_COLS.index(c) for c in TARGET_COLS]

print("Device:", DEVICE)
print("Number of features:", len(FEATURE_COLS))
print("Target columns:", TARGET_COLS)
```

```python
def inverse_transform_targets(values_scaled, scaler, target_indices):
    """
    Kthen parashikimet nga shkalla [0,1] në vlera reale USD
    vetëm për target columns.
    """
    values_scaled = np.asarray(values_scaled)
    output = np.zeros_like(values_scaled, dtype=float)

    for j, idx in enumerate(target_indices):
        data_min = scaler.data_min_[idx]
        data_max = scaler.data_max_[idx]
        output[:, j] = values_scaled[:, j] * (data_max - data_min) + data_min

    return output


def make_sequences_for_split(full_raw_df, full_scaled, target_start, target_end, lookback, horizon):
    """
    Krijon sekuenca me target brenda split-it të kërkuar.

    Input-i për çdo parashikim mbaron në origin_date.
    Target-i është origin_date + horizon.

    Kjo lejon që për Validation/Test të përdoret historia e mëparshme,
    por jo informacion nga e ardhmja.
    """
    X, y, meta = [], [], []

    target_start = pd.to_datetime(target_start)
    target_end = pd.to_datetime(target_end)

    for origin_idx in range(lookback - 1, len(full_raw_df) - horizon):
        target_idx = origin_idx + horizon

        target_date = full_raw_df.iloc[target_idx]["date"]
        origin_date = full_raw_df.iloc[origin_idx]["date"]

        if target_start <= target_date <= target_end:
            X.append(full_scaled[origin_idx - lookback + 1 : origin_idx + 1])
            y.append(full_scaled[target_idx, TARGET_IDXS])

            meta.append({
                "origin_idx": origin_idx,
                "target_idx": target_idx,
                "origin_date": origin_date,
                "date": target_date
            })

    X = np.array(X, dtype=np.float32)
    y = np.array(y, dtype=np.float32)
    meta_df = pd.DataFrame(meta)

    return X, y, meta_df


class CryptoSequenceDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]


class LSTMForecastModel(nn.Module):
    def __init__(self, input_size, hidden_size=128, num_layers=2, output_size=3, dropout=0.2):
        super().__init__()

        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )

        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        last_hidden = lstm_out[:, -1, :]
        out = self.dropout(last_hidden)
        return self.fc(out)


print("LSTM helper functions ready.")
```

```python
def train_lstm_for_horizon(horizon):
    print("\n" + "="*70)
    print(f"Training LSTM for horizon H={horizon}")
    print("="*70)

    # Scaler fit vetëm mbi Train
    scaler = MinMaxScaler()
    scaler.fit(train_df[FEATURE_COLS])

    full_scaled = scaler.transform(df[FEATURE_COLS])

    train_target_start = train_df["date"].iloc[horizon - 1]
    val_target_start = val_df["date"].iloc[horizon - 1]
    test_target_start = test_df["date"].iloc[horizon - 1]

    X_train, y_train, meta_train = make_sequences_for_split(
        full_raw_df=df,
        full_scaled=full_scaled,
        target_start=train_target_start,
        target_end=train_df["date"].max(),
        lookback=LOOKBACK,
        horizon=horizon
    )

    X_val, y_val, meta_val = make_sequences_for_split(
        full_raw_df=df,
        full_scaled=full_scaled,
        target_start=val_target_start,
        target_end=val_df["date"].max(),
        lookback=LOOKBACK,
        horizon=horizon
    )

    X_test, y_test, meta_test = make_sequences_for_split(
        full_raw_df=df,
        full_scaled=full_scaled,
        target_start=test_target_start,
        target_end=test_df["date"].max(),
        lookback=LOOKBACK,
        horizon=horizon
    )
    print("X_train:", X_train.shape, "y_train:", y_train.shape)
    print("X_val:  ", X_val.shape, "y_val:  ", y_val.shape)
    print("X_test: ", X_test.shape, "y_test: ", y_test.shape)

    train_loader = DataLoader(
        CryptoSequenceDataset(X_train, y_train),
        batch_size=BATCH_SIZE,
        shuffle=True,
        drop_last=False
    )

    val_loader = DataLoader(
        CryptoSequenceDataset(X_val, y_val),
        batch_size=BATCH_SIZE,
        shuffle=False,
        drop_last=False
    )

    test_loader = DataLoader(
        CryptoSequenceDataset(X_test, y_test),
        batch_size=BATCH_SIZE,
        shuffle=False,
        drop_last=False
    )

    model = LSTMForecastModel(
        input_size=len(FEATURE_COLS),
        hidden_size=HIDDEN_SIZE,
        num_layers=NUM_LAYERS,
        output_size=len(TARGET_COLS),
        dropout=DROPOUT
    ).to(DEVICE)

    optimizer = torch.optim.Adam(model.parameters(), lr=LR)
    criterion = nn.MSELoss()
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer,
        mode="min",
        patience=5,
        factor=0.5
    )

    best_val_loss = float("inf")
    best_state = None
    patience_counter = 0

    history = []

    for epoch in range(1, EPOCHS + 1):
        model.train()
        train_losses = []

        for xb, yb in train_loader:
            xb = xb.to(DEVICE)
            yb = yb.to(DEVICE)

            optimizer.zero_grad()
            preds = model(xb)
            loss = criterion(preds, yb)
            loss.backward()

            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()
            train_losses.append(loss.item())

        train_loss = float(np.mean(train_losses))

        model.eval()
        val_losses = []

        with torch.no_grad():
            for xb, yb in val_loader:
                xb = xb.to(DEVICE)
                yb = yb.to(DEVICE)

                preds = model(xb)
                loss = criterion(preds, yb)
                val_losses.append(loss.item())

        val_loss = float(np.mean(val_losses))
        scheduler.step(val_loss)

        history.append({
            "horizon": horizon,
            "epoch": epoch,
            "train_loss": train_loss,
            "val_loss": val_loss
        })

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_state = {k: v.detach().cpu().clone() for k, v in model.state_dict().items()}
            patience_counter = 0
        else:
            patience_counter += 1

        if epoch % 5 == 0 or epoch == 1:
            print(f"Epoch {epoch:03d} | Train Loss: {train_loss:.6f} | Val Loss: {val_loss:.6f}")

        if patience_counter >= PATIENCE:
            print(f"Early stopping at epoch {epoch}. Best val loss: {best_val_loss:.6f}")
            break

    model.load_state_dict(best_state)
    model.to(DEVICE)
    model.eval()

    all_preds_scaled = []

    with torch.no_grad():
        for xb, _ in test_loader:
            xb = xb.to(DEVICE)
            preds = model(xb).cpu().numpy()
            all_preds_scaled.append(preds)

    preds_scaled = np.vstack(all_preds_scaled)

    preds_usd = inverse_transform_targets(preds_scaled, scaler, TARGET_IDXS)
    actual_usd = inverse_transform_targets(y_test, scaler, TARGET_IDXS)

    return {
        "horizon": horizon,
        "model": model,
        "scaler": scaler,
        "history": pd.DataFrame(history),
        "meta_test": meta_test,
        "preds_usd": preds_usd,
        "actual_usd": actual_usd
    }


lstm_outputs = []

for horizon in HORIZONS:
    output = train_lstm_for_horizon(horizon)
    lstm_outputs.append(output)

print("All LSTM horizons completed.")
```

```python
lstm_predictions_rows = []
lstm_metrics_rows = []
lstm_history_all = []

for output in lstm_outputs:
    horizon = output["horizon"]
    meta_test = output["meta_test"]
    preds_usd = output["preds_usd"]
    actual_usd = output["actual_usd"]

    history_df = output["history"].copy()
    lstm_history_all.append(history_df)

    for asset_idx, asset in enumerate(["BTC", "ETH", "BNB"]):
        target_col = f"{asset}_close"

        asset_actual = actual_usd[:, asset_idx]
        asset_pred = preds_usd[:, asset_idx]

        origin_prices = []
        origin_dates = []
        target_dates = []

        for _, mrow in meta_test.iterrows():
            origin_idx = int(mrow["origin_idx"])
            origin_prices.append(float(df.iloc[origin_idx][target_col]))
            origin_dates.append(mrow["origin_date"])
            target_dates.append(mrow["date"])

        asset_pred_df = pd.DataFrame({
            "date": target_dates,
            "origin_date": origin_dates,
            "asset": asset,
            "horizon": horizon,
            "actual": asset_actual,
            "origin_price": origin_prices,
            "lstm_pred": asset_pred,
            "model": "LSTM"
        })

        metrics = compute_forecast_metrics(
            y_true=asset_pred_df["actual"],
            y_pred=asset_pred_df["lstm_pred"],
            origin_price=asset_pred_df["origin_price"]
        )

        metrics_row = {
            "model": "LSTM",
            "asset": asset,
            "horizon": horizon,
            "n_predictions": len(asset_pred_df),
            **metrics
        }

        lstm_predictions_rows.append(asset_pred_df)
        lstm_metrics_rows.append(metrics_row)

lstm_predictions_df = pd.concat(lstm_predictions_rows, ignore_index=True)
lstm_metrics_df = pd.DataFrame(lstm_metrics_rows)
lstm_history_df = pd.concat(lstm_history_all, ignore_index=True)

display(lstm_metrics_df)

lstm_predictions_df.to_csv(RESULTS_DIR / "lstm_predictions.csv", index=False)
lstm_metrics_df.to_csv(RESULTS_DIR / "lstm_metrics.csv", index=False)
lstm_history_df.to_csv(RESULTS_DIR / "lstm_training_history.csv", index=False)

print("LSTM predictions saved:", RESULTS_DIR / "lstm_predictions.csv")
print("LSTM metrics saved:", RESULTS_DIR / "lstm_metrics.csv")
print("LSTM history saved:", RESULTS_DIR / "lstm_training_history.csv")
```

```python
def build_tft_multiasset_return_dataframe(df):
    rows = []

    for asset in ["BTC", "ETH", "BNB"]:
        temp = pd.DataFrame()
        temp["date"] = df["date"]
        temp["asset"] = asset

        temp["close"] = df[f"{asset}_close"]
        temp["log_close"] = np.log(df[f"{asset}_close"])
        temp["target_return"] = temp["log_close"].diff()

        temp["open"] = df[f"{asset}_open"]
        temp["high"] = df[f"{asset}_high"]
        temp["low"] = df[f"{asset}_low"]
        temp["volume"] = df[f"{asset}_volume"]

        temp["RSI14"] = df[f"{asset}_RSI14"]
        temp["EMA7"] = df[f"{asset}_EMA7"]
        temp["EMA21"] = df[f"{asset}_EMA21"]
        temp["EMA50"] = df[f"{asset}_EMA50"]
        temp["MACD"] = df[f"{asset}_MACD"]
        temp["MACD_signal"] = df[f"{asset}_MACD_signal"]
        temp["BB_width"] = df[f"{asset}_BB_width"]

        temp["return1d"] = df[f"{asset}_return1d"]
        temp["vol7d"] = df[f"{asset}_vol7d"]

        temp["BTC_close_global"] = df["BTC_close"]
        temp["ETH_close_global"] = df["ETH_close"]
        temp["BNB_close_global"] = df["BNB_close"]

        temp["BTC_return_global"] = df["BTC_return1d"]
        temp["ETH_return_global"] = df["ETH_return1d"]
        temp["BNB_return_global"] = df["BNB_return1d"]

        temp["fear_greed"] = df["fear_greed"]
        temp["corr_BTC_ETH_30d"] = df["corr_BTC_ETH_30d"]
        temp["corr_BTC_BNB_30d"] = df["corr_BTC_BNB_30d"]
        temp["corr_ETH_BNB_30d"] = df["corr_ETH_BNB_30d"]

        temp["day_of_week"] = df["day_of_week"].astype(int)
        temp["month"] = df["month"].astype(int)
        temp["quarter"] = df["quarter"].astype(int)
        temp["is_weekend"] = df["is_weekend"].astype(int)

        rows.append(temp)

    out = pd.concat(rows, ignore_index=True)
    out["date"] = pd.to_datetime(out["date"])
    out = out.dropna().sort_values(["asset", "date"]).reset_index(drop=True)
    out["time_idx"] = (out["date"] - out["date"].min()).dt.days

    return out


tft_ret_df = build_tft_multiasset_return_dataframe(df)

print("TFT_LOG_RETURN dataset shape:", tft_ret_df.shape)
print("Date range:", tft_ret_df["date"].min().date(), "to", tft_ret_df["date"].max().date())
print("Rows per asset:")
display(tft_ret_df.groupby("asset").size())

tft_ret_df.to_csv(DATA_DIR / "tft_log_return_long_dataset.csv", index=False)
print("Saved:", DATA_DIR / "tft_log_return_long_dataset.csv")
```

```python
RET_KNOWN_REALS = [
    "time_idx",
    "day_of_week",
    "month",
    "quarter",
    "is_weekend"
]

RET_UNKNOWN_REALS = [
    "target_return",
    "open",
    "high",
    "low",
    "volume",
    "RSI14",
    "EMA7",
    "EMA21",
    "EMA50",
    "MACD",
    "MACD_signal",
    "BB_width",
    "return1d",
    "vol7d",
    "fear_greed",
    "corr_BTC_ETH_30d",
    "corr_BTC_BNB_30d",
    "corr_ETH_BNB_30d",
    "BTC_close_global",
    "ETH_close_global",
    "BNB_close_global",
    "BTC_return_global",
    "ETH_return_global",
    "BNB_return_global"
]

def get_ret_time_idx(date_value):
    date_value = pd.to_datetime(date_value)
    return int((date_value - tft_ret_df["date"].min()).days)

RET_TRAIN_END_IDX = get_ret_time_idx("2022-12-31")
RET_VAL_START_IDX = get_ret_time_idx("2023-01-01")
RET_VAL_END_IDX = get_ret_time_idx("2023-06-30")
RET_TEST_START_IDX = get_ret_time_idx("2023-07-01")
RET_TEST_END_IDX = get_ret_time_idx("2024-12-30")

def build_multiasset_return_tft_datasets(horizon):
    training = TimeSeriesDataSet(
        tft_ret_df[tft_ret_df["time_idx"] <= RET_TRAIN_END_IDX],
        time_idx="time_idx",
        target="target_return",
        group_ids=["asset"],
        min_encoder_length=60,
        max_encoder_length=60,
        min_prediction_length=horizon,
        max_prediction_length=horizon,
        static_categoricals=["asset"],
        time_varying_known_reals=RET_KNOWN_REALS,
        time_varying_unknown_reals=RET_UNKNOWN_REALS,
        target_normalizer=EncoderNormalizer(
            method="standard",
            transformation=None,
            max_length=60
        ),
        add_relative_time_idx=True,
        add_target_scales=True,
        add_encoder_length=True,
        allow_missing_timesteps=False
    )

    validation = TimeSeriesDataSet.from_dataset(
        training,
        tft_ret_df[tft_ret_df["time_idx"] <= RET_VAL_END_IDX],
        min_prediction_idx=RET_VAL_START_IDX,
        stop_randomization=True
    )

    testing = TimeSeriesDataSet.from_dataset(
        training,
        tft_ret_df[tft_ret_df["time_idx"] <= RET_TEST_END_IDX],
        min_prediction_idx=RET_TEST_START_IDX,
        stop_randomization=True
    )

    return training, validation, testing

print("TFT_LOG_RETURN dataset functions ready.")
```

```python
def train_multiasset_return_tft(horizon):
    print("\n" + "="*70)
    print(f"Training TFT_LOG_RETURN for H={horizon}")
    print("="*70)

    training, validation, testing = build_multiasset_return_tft_datasets(horizon)

    train_loader = training.to_dataloader(
        train=True,
        batch_size=64,
        num_workers=0
    )

    val_loader = validation.to_dataloader(
        train=False,
        batch_size=64,
        num_workers=0
    )

    test_loader = testing.to_dataloader(
        train=False,
        batch_size=64,
        num_workers=0
    )

    model = TemporalFusionTransformer.from_dataset(
        training,
        learning_rate=1e-3,
        hidden_size=64,
        attention_head_size=4,
        dropout=0.1,
        hidden_continuous_size=32,
        output_size=len(QUANTILES),
        loss=QuantileLoss(quantiles=QUANTILES),
        reduce_on_plateau_patience=4
    )

    print(f"TFT_LOG_RETURN parameters: {model.size()/1e3:.1f}k")

    early_stop_callback = EarlyStopping(
        monitor="val_loss",
        min_delta=1e-4,
        patience=8,
        verbose=True,
        mode="min"
    )

    trainer = pl.Trainer(
        max_epochs=50,
        accelerator="gpu" if torch.cuda.is_available() else "cpu",
        devices=1,
        gradient_clip_val=0.1,
        callbacks=[early_stop_callback],
        enable_progress_bar=True,
        enable_model_summary=False,
        logger=False
    )

    trainer.fit(
        model,
        train_dataloaders=train_loader,
        val_dataloaders=val_loader
    )

    return model, testing, test_loader

print("TFT_LOG_RETURN training function ready.")
```

```python
def extract_multiasset_return_tft_predictions_cumulative(model, testing_dataset, test_loader, horizon):
    prediction = model.predict(
        test_loader,
        mode="quantiles",
        return_x=True,
        trainer_kwargs={
            "accelerator": "gpu" if torch.cuda.is_available() else "cpu",
            "devices": 1
        }
    )

    output = prediction.output
    if hasattr(output, "detach"):
        output = output.detach().cpu().numpy()

    p50_return = output[:, :, 1]

    x = prediction.x
    decoder_time_idx = x["decoder_time_idx"].detach().cpu().numpy()

    target_time_idx = decoder_time_idx[:, -1]
    origin_time_idx = target_time_idx - horizon

    index_df = testing_dataset.x_to_index(x)

    records = []

    for i in range(len(index_df)):
        asset = str(index_df.iloc[i]["asset"])

        target_idx = int(target_time_idx[i])
        origin_idx = int(origin_time_idx[i])

        actual_row = tft_ret_df[
            (tft_ret_df["asset"] == asset) &
            (tft_ret_df["time_idx"] == target_idx)
        ]

        origin_row = tft_ret_df[
            (tft_ret_df["asset"] == asset) &
            (tft_ret_df["time_idx"] == origin_idx)
        ]

        if actual_row.empty or origin_row.empty:
            continue

        origin_price = float(origin_row["close"].iloc[0])
        actual_price = float(actual_row["close"].iloc[0])

        predicted_cumulative_return = float(np.sum(p50_return[i, :]))
        predicted_price = origin_price * np.exp(predicted_cumulative_return)

        actual_cumulative_return = float(np.log(actual_price / origin_price))

        records.append({
            "date": actual_row["date"].iloc[0],
            "origin_date": origin_row["date"].iloc[0],
            "asset": asset,
            "horizon": horizon,
            "actual": actual_price,
            "origin_price": origin_price,
            "actual_cumulative_return": actual_cumulative_return,
            "predicted_cumulative_return": predicted_cumulative_return,
            "tft_pred": predicted_price,
            "model": "TFT_LOG_RETURN"
        })

    return pd.DataFrame(records)


def evaluate_tft_log_return_predictions(pred_df, horizon):
    rows = []

    for asset in ["BTC", "ETH", "BNB"]:
        temp = pred_df[pred_df["asset"] == asset].copy()

        metrics = compute_forecast_metrics(
            y_true=temp["actual"],
            y_pred=temp["tft_pred"],
            origin_price=temp["origin_price"]
        )

        rows.append({
            "model": "TFT_LOG_RETURN",
            "asset": asset,
            "horizon": horizon,
            "n_predictions": len(temp),
            **metrics
        })

    return pd.DataFrame(rows)

print("TFT_LOG_RETURN prediction functions ready.")
```

```python
# ============================================================
# Diebold-Mariano Test + Wilcoxon Signed-Rank Test
# ============================================================

from scipy import stats

def diebold_mariano_test(e1, e2, horizon=1, power=2):
    """
    e1 = actual - prediction_model_a
    e2 = actual - prediction_model_b

    Negative DM => model_a has lower loss than model_b.
    Positive DM => model_b has lower loss than model_a.
    """
    e1 = np.asarray(e1, dtype=float)
    e2 = np.asarray(e2, dtype=float)

    if power == 2:
        d = e1**2 - e2**2
    elif power == 1:
        d = np.abs(e1) - np.abs(e2)
    else:
        raise ValueError("power must be 1 or 2")

    d = d[~np.isnan(d)]
    n = len(d)

    if n < 5:
        return np.nan, np.nan

    d_mean = np.mean(d)

    lag = max(horizon - 1, 0)
    gamma0 = np.var(d, ddof=1)
    var_d = gamma0

    for l in range(1, lag + 1):
        cov = np.cov(d[l:], d[:-l], ddof=1)[0, 1]
        weight = 1 - l / (lag + 1)
        var_d += 2 * weight * cov

    if var_d <= 0 or np.isnan(var_d):
        return np.nan, np.nan

    dm_stat = d_mean / np.sqrt(var_d / n)
    p_value = 2 * (1 - stats.norm.cdf(abs(dm_stat)))

    return float(dm_stat), float(p_value)


def run_statistical_tests(pred_df):
    rows = []

    model_pairs = [
        ("TFT_LOG_RETURN", "ARIMA", "tft_pred", "arima_pred"),
        ("TFT_LOG_RETURN", "LSTM", "tft_pred", "lstm_pred"),
        ("LSTM", "ARIMA", "lstm_pred", "arima_pred"),
    ]

    for asset in ["BTC", "ETH", "BNB"]:
        for horizon in [1, 7, 30]:
            temp = pred_df[
                (pred_df["asset"] == asset) &
                (pred_df["horizon"] == horizon)
            ].copy()

            actual = temp["actual"].values

            for model_a, model_b, pred_a_col, pred_b_col in model_pairs:
                pred_a = temp[pred_a_col].values
                pred_b = temp[pred_b_col].values

                e_a = actual - pred_a
                e_b = actual - pred_b

                dm_se, dm_se_p = diebold_mariano_test(
                    e1=e_a,
                    e2=e_b,
                    horizon=horizon,
                    power=2
                )

                dm_ae, dm_ae_p = diebold_mariano_test(
                    e1=e_a,
                    e2=e_b,
                    horizon=horizon,
                    power=1
                )

                try:
                    wilcoxon_stat, wilcoxon_p = stats.wilcoxon(
                        np.abs(e_a),
                        np.abs(e_b),
                        zero_method="wilcox",
                        alternative="two-sided"
                    )
                except Exception:
                    wilcoxon_stat, wilcoxon_p = np.nan, np.nan

                rows.append({
                    "asset": asset,
                    "horizon": horizon,
                    "model_a": model_a,
                    "model_b": model_b,
                    "dm_squared_error_stat": round(dm_se, 4) if not np.isnan(dm_se) else np.nan,
                    "dm_squared_error_p_value": round(dm_se_p, 6) if not np.isnan(dm_se_p) else np.nan,
                    "dm_absolute_error_stat": round(dm_ae, 4) if not np.isnan(dm_ae) else np.nan,
                    "dm_absolute_error_p_value": round(dm_ae_p, 6) if not np.isnan(dm_ae_p) else np.nan,
                    "wilcoxon_stat": round(float(wilcoxon_stat), 4) if not np.isnan(wilcoxon_stat) else np.nan,
                    "wilcoxon_p_value": round(float(wilcoxon_p), 6) if not np.isnan(wilcoxon_p) else np.nan,
                    "n": len(temp),
                    "interpretation": "Negative DM means model_a has lower loss than model_b"
                })

    return pd.DataFrame(rows)


stat_tests_df = run_statistical_tests(combined_predictions_df)

display(stat_tests_df)

stat_tests_df.to_csv(
    TESTS_DIR / "statistical_tests_dm_wilcoxon.csv",
    index=False
)

print("Saved:", TESTS_DIR / "statistical_tests_dm_wilcoxon.csv")
print("Exists:", (TESTS_DIR / "statistical_tests_dm_wilcoxon.csv").exists())
```

```python
# ============================================================
# Clean statistical interpretation summary
# ============================================================

def interpret_dm(row):
    p = row["dm_squared_error_p_value"]
    dm = row["dm_squared_error_stat"]
    model_a = row["model_a"]
    model_b = row["model_b"]

    if pd.isna(p):
        return "No conclusion"

    if p >= 0.05:
        return f"No statistically significant difference between {model_a} and {model_b}"

    if dm < 0:
        return f"{model_a} significantly outperforms {model_b}"

    if dm > 0:
        return f"{model_b} significantly outperforms {model_a}"

    return "No conclusion"


stat_summary_df = stat_tests_df.copy()
stat_summary_df["conclusion_squared_error"] = stat_summary_df.apply(interpret_dm, axis=1)

stat_summary_clean_df = stat_summary_df[
    [
        "asset",
        "horizon",
        "model_a",
        "model_b",
        "dm_squared_error_stat",
        "dm_squared_error_p_value",
        "wilcoxon_p_value",
        "n",
        "conclusion_squared_error"
    ]
].copy()

display(stat_summary_clean_df)

stat_summary_clean_df.to_csv(
    TESTS_DIR / "statistical_tests_clean_summary.csv",
    index=False
)

print("Saved:", TESTS_DIR / "statistical_tests_clean_summary.csv")
print("Exists:", (TESTS_DIR / "statistical_tests_clean_summary.csv").exists())
```
